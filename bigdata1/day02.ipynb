#### 3.2.1.3 세트

세트는 수학의 집합을 구현한 자료형.  
세트에 저장된 값들은 순서가 없기 때문에 인덱싱과 슬라이싱 불가.  
대신 중복 값 저장이 불가능하다는 특징을 활용해 중복 제거용으로 사용하거나  
교집합, 합집합, 차집합과 같은 집합 개념이 필요한 경우 사용

>사용법
* {값1, 값2}
* set(반복가능객체)
<br>

다만 빈 세트 사용 시에는 중괄호를 사용할 수 없음. (이는 딕셔너리)
빈 세트 생성 시 set 함수 이용

```python
# Jaccard Similarity
string1 = '나는 파이썬을 좋아합니다'
string2 = '나는 파이썬을 정말 좋아합니다'

splitted_string1 = string1.split()
splitted_string2 = string2.split()
print(splitted_string1)
print(splitted_string2)

splitted_string1_set = set(splitted_string1)
splitted_string2_set = set(splitted_string2)

print(splitted_string1_set.intersection(splitted_string2_set))
print(splitted_string1_set.union(splitted_string2_set))

print(
    len(splitted_string1_set.intersection(splitted_string2_set)) 
    / len(splitted_string1_set.union(splitted_string2_set))
)
```

#### 3.2.1.4 딕셔너리

딕셔너리는 말 그대로 사전을 의미.  
파이썬에서는 '키'와 '값'을 '단어'와 '단어의 의미'처럼 사용.  



> 사용법
1. dict = {키1: 값1, 키2: 값2, ...}
2. dict = dict(키1=값1, 키2=값2)
3. dict = dict([(키1, 값1), (키2, 값2)]) 
4. dict = dict({키1: 값1, 키2: 값2}) 
5. dict = dict(zip([키1, 키2], [값1, 값2]))

```python
print({'이름': '홍길동', '나이': 16, '직업': '학생'})
print(dict(이름='홍길동', 나이=16, 직업='학생'))
print(dict([('이름', '홍길동'), ('나이', 16), ('직업', '학생')]))
print(dict({'이름': '홍길동', '나이': 16, '직업': '학생'}))
print(dict(zip(['이름', '나이', '직업'], ['홍길동', 16, '학생'])))
```

### 3.3.2 while

1.   while  
while문은 특정 조건을 만족하는 동안 반복해서 수행해야 하는 코드를 작성할 때 사용.  

> * while문의 기본 구조는 다음와 같음  
while 조건식:  
     &nbsp;&nbsp;&nbsp;&nbsp;반복실행문.

<br>

2.   while문이 필요한 경우  
while문은 주로 횟수가 정해지지 않은 연산을 진행할 때나  
어떤 특정 동작이 지속될 때 사용
<br>


3.   while문의 중첩  
while문 내부에 또 다른 while문이 나타나는 것을 while문의 중첩이라고 함
<br>
반복처리해야 하는 대상이 2개 이상이면 while문의 중첩이 필요.  

```python
# 1~5까지 더하기
total = 0
i = 1
while i <= 5:
    total += i
    i += 1

print(total)
```

```python
# hello world 10번 출력
i = 1
while i <= 10:
    print(f'{i}: hello world')
    i += 1
```

```python
# 구구단
for i in range(2, 10):
    for j in range(1, 10):
        print(f'{i} * {j} = {i*j}')
    print()
```

```python
i = 2

while i < 10:
    j = 1
    while j < 10:
        print(f'{i} * {j} = {i*j}')
        j += 1
    print()
    i += 1
```

```python
# 비밀번호 맞추기
while True:
    password = int(input('비밀번호를 입력해주세요 >> '))

    if password == 486:
        print('정답')
        break #맞으면 반복 중단
    else:
        print('오답') #틀렸을 경우 while문 계속해서 반복
```

**가변 매개변수**
함수로 전달해야 하는 인수의 개수가 정해지지 않아도  
매개변수 선언 가능한데 이를 가변 매개변수라고 함  

가변 매개변수를 만드는 키워드는 애스터리스크(*)이며 개수에 따른 차이가 있음

1. def show(*args):  
&nbsp;&nbsp;&nbsp;&nbsp;for item in args:  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;print(item)  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;show('happy', 'birthday')  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;-> 인자를 튜플로 전달

2. def show(**kwargs):  
&nbsp;&nbsp;&nbsp;&nbsp;for key, item in args:  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;print(key, item)

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;show(a=1, b=2)  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;-> 인자를 딕셔너리로 전달

```python
def calculate_score(score):
    if score < 0:
        return -1
    if score > 100:
        return -1

    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'
```

```python
calculate_score(101)
```

```python
def average(*args): #인자로 들어가는 것들은 숫자만 가능
    print(args)
    print(f'average: {sum(args)/len(args)}')
```

```python
average(90, 80, 70, 60, 50, 50, 50, 50, 50, 50, 50, 50, 50, 50, 50, 50, 50)
```

```python
def average(**kwargs): #딕셔너리로 (키=밸류)
    print(kwargs)
    total = 0
    for subject, score in kwargs.items():
        total += score

    print(f'average: {total/len(kwargs)}')

# 변수명 바꿔도 상관없음 (마음대로 가능)
# 하지만 가변 인자에 대해 *별표 하나는 args로 
# **별표 두 개는 **kwdargs로 쓰기로 암묵적인 약속
# 그래서 이것만 지켜주기!
```

```python
average(korean=90, english=80, math=70, society=60, science=50, giga=50, art=60)
```
